```python
import numpy as np
import matplotlib.pyplot as plt

# 这个设置才可以让Mac下可以显示中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

plt.rcParams['figure.figsize'] = (15.0,10.0) # 设置图像的显示区域
```

```python
x_train = np.array([np.random.randint(-10,10) for _ in range(5000)])
x_train
```

```python
y_train = np.array([1 if x == 0 else np.sin(x) for x in x_train])
y_train
```

```python
y_train
```

```python
plt.scatter(x_train,y_train,color='b')
plt.xlabel("特征x1")
plt.ylabel("特征x2")
plt.show()
```

```python
#生成训练集数据（xi,y）
import matplotlib.pyplot as plt


np.random.seed(2020)
train_x = np.random.uniform(low = -10,high=10,size=5000)
train_x = np.round(train_x,4)
train_y1=[(np.sin(i)/i) if i !=0 else 1 for i in train_x]
noise = np.random.uniform(low = -0.2,high=0.2,size=5000)
train_y2 = train_y1 + noise
# 
plt.scatter(train_x,train_y1)
plt.scatter(train_x,train_y2)
plt.show()
```
