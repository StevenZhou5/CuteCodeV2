```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.linspace(0,100,100)
```

```python
X_train
```

```python
def getY(x):
    if x < 34:
        return 100 - x;
    elif x < 68:
        return 67
    else:
        return 100 - x
Y_train = np.array(list(map(getY,X_train)))
Y_train
```

```python
plt.title("Raw data trend charts") # 原始数据趋势图
plt.xlabel("Vehicle mileage") # 车辆行驶里程
plt.ylabel("Residual value of the vehicle") # 车辆的残值
plt.plot(X_train,Y_train,)
plt.show()
```

```python
# 建立线性拟合模型
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(8,input_shape =(1,),activation='relu'),
    tf.keras.layers.Dense(16,activation='relu'),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(16,activation='relu'),
    tf.keras.layers.Dense(1)
])
model.summary()

model.compile(loss = 'mse',optimizer = tf.keras.optimizers.Adam(learning_rate=0.03),metrics=[['accuracy', 'mse']])
history = model.fit(X_train,Y_train,epochs=1000)
```

```python
# 对比预测结果和真实结果的曲线
Y_predict = model.predict(X_train)
Y_predict = Y_predict.reshape(-1)
Y_predict
plt.plot(X_train,Y_train,color ='red',label='train')
plt.plot(X_train,Y_predict,color = 'blue',label='predict')
plt.legend()
plt.show()
```

```python
# 如果用取值间隔很小的点来绘制
X_more = np.linspace(0,100,1000)
Y_more = model.predict(X_more).reshape(-1)
plt.plot(X_train,Y_train,color='red',label = 'train')
plt.plot(X_more,Y_more,color='blue',label = 'more')
plt.legend()
plt.show()
```

```python
# 使用X_more与Y_more来进行拐点搜索
def get_change_list(X,Y):
    k1 = (Y[1] - Y[0])/(X[1] - X[0])
#     print(last)
    X_change_list = [] # 记录拐点值
    for i in range(1,len(X)-1):
        k2 = (Y[i+1] - Y[i])/(X[i+1] - X[i])
        dk = abs((k2 - k1)/k1)
#         print(dk)
        k1 = k2
        if dk > 0.5:
#             print('x = %s 是一个拐点'%(X[i]))
            X_change_list.append(X[i])
    return X_change_list
X_change_list = get_change_list(X_more,Y_more)
X_change_list
```

```python
# 使用拐点的真实值去修正模型
X_change_list_che300_sample = np.array(X_change_list[4:6])
print(X_change_list_che300_sample)
Y_change_list_che300_sample = 100 - X_change_list_che300_sample
Y_change_list_che300_sample
```

```python
# 第一种做法，直接将拐点数据加入训练集
X_train = np.vstack((X_train.reshape(-1,1),X_change_list_che300_sample.reshape(-1,1)))
Y_train = np.vstack((Y_train.reshape(-1,1),Y_change_list_che300_sample.reshape(-1,1)))
X_train.shape
```

```python
model.fit(X_train,Y_train,epochs=1000)
```

```python
# 验证修正后的效果的改变
X_more = np.linspace(0,100,1000)
Y_more = model.predict(X_more).reshape(-1)
plt.plot(X_train,Y_train,color='red',label = 'train')
plt.plot(X_more,Y_more,color='blue',label = 'more')
plt.legend()
plt.show()
```

```python
# 除了添加修正数据外，还将原始有偏差的拐点附件的数据取出掉
X_train = np.linspace(0,100,100)
def getY(x):
    if x < 34:
        return 100 - x;
    elif x < 68:
        return 67
    else:
        return 100 - x
Y_train = np.array(list(map(getY,X_train)))
X_train = X_train.reshape(-1,1)
Y_train = Y_train.reshape(-1,1)
print(X_train.shape)
print(Y_train.shape)
```

```python
delete_items = []
for i in range(len(X_train)):
    x_i = X_train[i]
    for x_change_point in X_change_list_che300_sample: # 移除和车300结果有误差的拐点的数据
        dx = abs(x_change_point - X_train[i])
#         print('第 %f 个值和 %s 的差距为 %s'%(i,x_change_point,dx))
        if dx < 10:
#             print('删除了第 %d 个值'%(i))
            delete_items.append(i)
X_train = np.delete(X_train,delete_items,axis=0)
print(X_train.shape)
Y_train = np.delete(Y_train,delete_items,axis=0)
print(Y_train.shape)
```

```python
X_train = np.vstack((X_train,X_change_list_che300_sample.reshape(-1,1)))
Y_train = np.vstack((Y_train,Y_change_list_che300_sample.reshape(-1,1)))
X_train.shape
```

```python
model.fit(X_train,Y_train,epochs=1000)
```

```python
# 验证修正后的效果的改变
X_more = np.linspace(0,100,1000)
Y_more = model.predict(X_more).reshape(-1)
plt.plot(X_train,Y_train,color='red',label = 'train')
plt.plot(X_more,Y_more,color='blue',label = 'more')
plt.legend()
plt.show()
```
