```python
import tensorflow as tf
import numpy as np
import pandas as pd

pd.set_option("display.max_columns",100)
pd.set_option("display.max_rows",100)
```

```python
X = np.array([np.random.randint(-10,100) for _ in range(1000)])
print(X)
Y = np.array([(2*x if x > 0 else 3*x)for x in X]) # 故意在4的时候让值固定，
print(Y)
```

```python
sample_weights = np.array([(1.0 if x>0 else 2.0) for x in X])
print(sample_weights)
```

```python
input_layers = tf.keras.layers.Input(shape=(1,),name="x")
hidden_layers = tf.keras.layers.Dense(32)(input_layers)
output_layers = tf.keras.layers.Dense(1)(hidden_layers)

model = tf.keras.models.Model(inputs=[input_layers],
                              outputs=[output_layers]
                             )
model.summary()

model.compile(tf.keras.optimizers.Adam(0.03),'mae')
history = model.fit({"x":X},Y,epochs=1000,sample_weight=None)
```

```python
model.predict([3])
```

```python
# 未加权
model.predict([-3])
```

```python
input_layers = tf.keras.layers.Input(shape=(1,),name="x")
input_layers2 = tf.keras.layers.Input(shape=(1,),name="x2")
hidden_layers1 = tf.keras.layers.Dense(16)(input_layers)
hidden_layers2 = tf.keras.layers.Dense(32)(hidden_layers1)
hidden_layers3 = tf.keras.layers.Dense(16)(hidden_layers2)
output_layers = tf.keras.layers.Dense(1)(hidden_layers3)

model = tf.keras.models.Model(inputs=[input_layers,input_layers2],
                              outputs=[output_layers]
                             )
model.summary()

model.compile(tf.keras.optimizers.Adam(0.03),'mae')
history = model.fit({"x":X,"x2":X},Y.reshape((-1,1)),epochs=1000,sample_weight=sample_weights)
```

```python
model.predict([3])
# 5倍加权：array([[3.776136]], dtype=float32)
# 2倍加权：array([[5.560416]], dtype=float32)
```

```python
# 加权之后的结果，加权过大对其他样本影响较大，加权过小并没有明显影响
# 我们使用的是连续特征做测试，其实是不太合理的，如果是离散特征，加权的效果可能会好一些；
model.predict([-3])
# 5倍加权：array([[-8.722444]], dtype=float32)
# 2倍加权：array([[-6.5836954]], dtype=float32)
```
