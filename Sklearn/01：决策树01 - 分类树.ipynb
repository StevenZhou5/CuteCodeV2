```python
# 建模
clf = tree.DecisionTreeClassifier(criterion='entropy',
                                  random_state=0, # 防止每次运行结果不一致
                                  splitter='random', # 默认是'best':用于控制训练集合验证集的分割方式
                                  min_samples_leaf=3, # 每个叶子节点最少的样本数
                                  min_samples_split=5, # 一个结点至少包含min_samples_split个样本，否侧不允许分支
                                  max_depth=3, # 数的最大深度,在数据量少的时候可以防止过拟合
                                 ) # 建立决策树
clf = clf.fit(Xtrain,Ytrain) # 模型训练
score = clf.score(Xtest,Ytest) # 打分
print(score)

feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','青花素','颜色强度',
                '色调','od280/od315稀释葡萄酒','脯氨酸']
# 画决策树
import graphviz
dot_data = tree.export_graphviz(clf,
                                feature_names = feature_name,
                                class_names = ["琴酒",'雪莉','贝尔摩德'],
                                filled=True, # 是否填充颜色
                                rounded=True # 圆角
                               )
graph = graphviz.Source(dot_data)
graph
```

# 如果确定超参数的值，使用超参数曲线

```python
# 绘画超参数曲线，(超参数取值，模型效果)
import matplotlib.pyplot as plt
test = []
for i in range(1,11,1):
    clf = tree.DecisionTreeClassifier(criterion='entropy',
                                      random_state=0, # 防止每次运行结果不一致
                                      splitter='random', # 默认是'best':用于控制训练集合验证集的分割方式
                                      min_samples_leaf=3, # 每个叶子节点最少的样本数
                                      max_depth=i, # 数的最大深度,在数据量少的时候可以防止过拟合
                                     ) # 建立决策树
    clf = clf.fit(Xtrain,Ytrain) # 模型训练
    score = clf.score(Xtest,Ytest) # 打分
#     print(score)
    test.append(score)
plt.plot(range(1,11),test,color='red',label='max_depth')
plt.legend()
plt.show()
```
