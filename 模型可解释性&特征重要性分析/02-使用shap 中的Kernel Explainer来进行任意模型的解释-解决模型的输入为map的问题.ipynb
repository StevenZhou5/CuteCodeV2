```python
import tensorflow as tf
import numpy as np
import shap
import time
```

```python
x1 = np.array([np.random.randint(0,100) for _ in range(10000)])
x2 = np.array([np.random.randint(0,100) for _ in range(10000)])
x3 = np.array([np.random.randint(0,100) for _ in range(10000)])
y = x1*3 + x2*0.5 + x3*0
```

```python
input1 = tf.keras.layers.Input(shape=(1,),name="x1")
input2 = tf.keras.layers.Input(shape=(1,),name="x2")
input3 = tf.keras.layers.Input(shape=(1,),name="x3")

concatenate = tf.keras.layers.concatenate(inputs=[input1,input2,input3],axis=-1)
output = tf.keras.layers.Dense(1,activation='relu')(concatenate)

model = tf.keras.models.Model(inputs=[input1,input2,input3],
                              outputs=[output])
model.summary()

model.compile(tf.keras.optimizers.Adam(0.03),
              'mse')
```

```python
history = model.fit({"x1":x1,"x2":x2,"x3":x3},
                    y,
                    batch_size=256,
                    epochs=100
                   )
```

```python
test_x1 = np.array([np.random.randint(0,100) for _ in range(100)])
test_x2 = np.array([np.random.randint(0,100) for _ in range(100)])
test_x3 = np.array([np.random.randint(0,100) for _ in range(100)])
print(test_x1,"\n",test_x2,"\n",test_x3)
```

```python
test_predict1 = model.predict([test_x1, test_x2,test_x3])
print(test_predict1.reshape((-1)))
```

```python
test_predict2 = model.predict({"x1":test_x1,"x2":test_x2,"x3":test_x3})
print(test_predict2.reshape(-1))
```

# 使用SHAP库 的Kernel Explainer进行特征贡献度的解释

```python
def my_predict(data):
    tem_x1 = data.T[0]
    tem_x2 = data.T[1]
    tem_x3 = data.T[2]
    return model.predict({"x1":tem_x1,"x2":tem_x2,"x3":tem_x3})
```

```python
test_x_expand = np.expand_dims((test_x1,test_x2,test_x3),axis=0).T.reshape((-1,3))
my_predict(test_x_expand).reshape(-1)
```

```python
train_x_expand = np.expand_dims((x1,x2,x3),axis=0)
train_x_expand = train_x_expand.T.reshape((-1,3))
train_x_expand
```

```python
explainerKNN =  shap.KernelExplainer(my_predict,train_x_expand)
explainerKNN.expected_value
```

```python
shap_values_KNN_test = explainerKNN.shap_values(test_x_expand)
```

```python
sum(shap_values_KNN_test[0][0]) + float(explainerKNN.expected_value)
```
