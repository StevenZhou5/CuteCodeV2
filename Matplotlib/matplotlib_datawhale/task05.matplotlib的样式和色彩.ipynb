###  一、matplotilb的绘图样式（style）

绘图样式指的是绘图的整体风格，可以在绘制的时候就指定元素的样式，也可以使用一种整体的主题风格。

matplotlib库还提供了四种批量修改全局样式的方式：

* **预设样式** 套模板 plt.style.use('')
* **用户自定义样式** 创建.mplstyle文件
* **修改默认rc设置**  修改matplotlib.rcParams设置

********
#### 1.1.matplotlib预先定义样式

```python
plt.figure(figsize=(4,4))
plt.style.use('default')
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.figure(figsize=(4,4))
plt.style.use('ggplot')
plt.plot(x,y)
```

```python
# 更多样式,

print(plt.style.available)
```


#### 1.2.用户自定义stylesheet

本质官方的style文档也是.mplstyle的文件，可以在..\Anaconda3\Lib\site-packages\matplotlib\mpl-data\stylelib文件夹下看到官方预设的style的文件




因此我们可以通过自定义创建后缀名为.mplstyle的文件，然后再用plt.style.use(),
文件设置的规则可参照 [官方文档](https://github.com/matplotlib/matplotlib/blob/master/matplotlibrc.template)

现在我们新建一个mystyle.mplstyl，文件中编辑样式内容：

>axes.titlesize:24   
>axes.labelsize:20  
>lines.linewidth:3  
>lines.marksize:10  
>xtick.labelsize:16  
>ytick.labelsize:16  

```python
#
plt.figure(figsize=(4,4))
plt.style.use('mystyle.mplstyle')
plt.plot(x,y)
```

值得特别注意的是，matplotlib支持混合样式的引用，只需在引用时输入一个样式列表，若是几个样式中涉及到同一个参数，右边的样式表会覆盖左边的值。

```python
# 
plt.figure(figsize=(4,4))
plt.style.use('dark_background')
plt.plot(x,y)
```

```python
plt.figure(figsize=(4,4))
plt.style.use(['dark_background','mystyle.mplstyle'])
plt.plot(x,y)
```

#### 1.3.设置rcparams

matplotlib所有的默认设置都保存到 matplotlib.rcParams的变量中。

可以通过以下三种方式**查看默认设置**：

* <font color = red>mpl.rcParams</font>
* <font color = blue>mpl.rcParamsDefault()</font>
* <font color = green>mpl.rc_params()</font>

同时我们可以通过修改matplotlib.rcParams中的默认设置来修改绘图的样式；

* 修改单个样式 mpl.rcParams['lines.linewidth'] = 2
* 同时修改多个样式 mpl.rc('lines', linewidth=4, linestyle='-.')

```python
# 
mpl.rcParams
```

```python
# 
plt.figure(figsize=(3,2))
plt.style.use('mystyle.mplstyle')
plt.plot(x,y)
```

```python
#
plt.figure(figsize=(4,4))
mpl.rcParams['lines.linewidth']=1
plt.plot(x,y)
```

```python
plt.figure(figsize=(4,4))
mpl.rc('lines',linewidth=5, linestyle='-.')
plt.plot(x,y)
```

### 二、matplotlib的色彩设置（color）

关于配色的详情不详细展开，datawhale提供的扩展材料：

matplotlib中色彩设置的方式有以下几种方式：

* **RGB或RGBA**  color=(0.1, 0.2, 0.5,O.5) 颜色用0-1之间的浮点数表示，四个分量按顺序分别为(red, green, blue, alpha)
* **HEX RGB 或 RGBA**  color='#0f0f0f80' 用十六进制颜色码表示，同样最后两位表示透明度，可省略
* **灰度色阶**  color = 0.5 当只有一个位于[0,1]的值时，表示灰度色阶
* **单字符基本颜色**  color='m' matplotlib有八个基本颜色，可以用单字符串来表示，分别是'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'，对应的是blue, green, red, cyan, magenta, yellow, black, and white的英文缩写
* **使用颜色名字** color='darkorange' 每个颜色对应的名称
* **colormap设置一组颜色**

```python
plt.style.use('default')

font = {'family': 'Times New Roman',
        'color':  'black',
        'weight': 'normal',
        'size': 10,
        }
```

```python
# RGB或RGBA
plt.figure(figsize=(4,3))
plt.plot(x,np.cos(x),color = (0.2,0.3,0.4))
plt.text(3,-0.8,'y=cos(x)',fontdict=font)
plt.plot(x,y,color=(0.2,0.3,0.4,0.5))
plt.text(8,1,'y=sin(x)',fontdict=font)
```

```python
# HEX RGB 或 RGBA
plt.figure(figsize=(4,3))
plt.plot(x,y,color='#0f0f0f80')
```

```python
# 灰度图
plt.figure(figsize=(4,3))
plt.plot(x,y,color='0.5')
```

```python
# 字符本身名称


plt.figure(figsize=(4,3))
plt.plot(x,np.cos(x),color = 'c')
plt.text(1.5,-0.5,'y=cos(x)',fontdict=font)

# 颜色名称
plt.plot(x,y,color='darkorange')
plt.text(8,1,'y=sin(x)',fontdict=font)
```

**colormap设置一组颜色**

有些图表支持 colormap 的方式配置一组颜色，

colormap有五种类型：

* **顺序（Sequential)** 通常使用单一色调，逐渐改变亮度和颜色渐渐增加，用于表示有顺序的信息
* **发散（Diverging）** 改变两种不同颜色的亮度和饱和度，这些颜色在中间以不饱和的颜色相遇；当绘制的信息具有关键中间值（例如地形）或数据偏离零时，应使用此值。
* **循环（Cyclic)** 改变两种不同颜色的亮度，在中间和开始/结束时以不饱和的颜色相遇。用于在端点处环绕的值，例如相角，风向或一天中的时间
* **定性（Qualitative）** 常是杂色，用来表示没有排序或关系的信息
* **杂色（Miscellaneous)** 一些在特定场景使用的杂色组合，如彩虹，海洋，地形等。
