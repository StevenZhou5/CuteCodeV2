###  一、Figure和Axes上的文本**



Matplotlib绘图方式有两种，一种是通过pyplot中的绘图函数直接绘图，一种是通过创建对象绘制图形，这两种方式创建文本的方式略有不同，

下面的命令是介绍了通过pyplot API和objected-oriented API分别创建文本的方式。




#### 1.1.text

* pyplot API:**plt.text**(self, x, y, s, fontdict=None, **kwargs)
    * 参数说明： 
    * s : 要添加的文本
    * xy：文本坐标
    * fontdict:此参数是一个可选参数，并且是一个覆盖默认文本属性的字典 即透明度，颜色等等属性
            具体的参数说明参照[官方网站](https://matplotlib.org/api/text_api.html?highlight=text#matplotlib.text.Text)
* OO API:**Axes.text**(self, x, y, s, fontdict=None, **kwargs)
    * 参数同上


```python
## 文字风格
font1 = {'family':'Times New Roman',
        'color':'purple',
        'weight':'normal',
        'size':16}

font2 = {'family': 'Times New Roman',
        'color':  'lightblue',
        'weight': 'normal',
        'size': 16,
        }
font3 = {'family': 'serif',
        'color':  'blue',
        'weight': 'bold',
        'size': 14,
         'alpha':0.4
        }
font4 = {'family': 'Calibri',
        'color':  'navy',
        'weight': 'normal',
        'size': 17,
         'alpha':0.2
        }
```

```python

x = np.linspace(0.0,5.0,100)
y = np.cos(2*np.pi*x)*np.exp(-x/3)

#
plt.plot(x,y,'--',color ='darkorange')
plt.title('Damped exponential decay', fontdict=font1)


# 
plt.text(2,0.65,r'$\cos(2 \pi x) \exp(-x/3)$', fontdict=font2)


#
plt.xlabel('Y=time (s)', fontdict=font3)
plt.ylabel('X=voltage(mv)', fontdict=font4)
```

#### 1.2.title和set_title

* pyplot API:**matplotlib.pyplot.title**(label, fontdict=None, loc=None, pad=None, *, y=None, ** kwargs)
    * 参数说明： 
    * label：str，此参数是要添加的文本
    * fontdict：dict，此参数是控制title文本的外观,
    * loc: str，{‘center’, ‘left’, ‘right’}默认为center
    * pad: float,该参数是指标题偏离图表顶部的距离，默认为6。
    * y: float，该参数是title所在axes垂向的位置。默认值为1，即title位于axes的顶部。
* OO API:**Axes.set_title**(self, label, fontdict=None, loc=None, pad=None, *, y=None, **kwargs)
    * 参数同上

```python
# fontdict的默认参数

'''
{'fontsize': rcParams['axes.titlesize'],
 'fontweight': rcParams['axes.titleweight'],
 'color': rcParams['axes.titlecolor'],
 'verticalalignment': 'baseline',
 'horizontalalignment': loc}
 
 '''
```

```python
x = np.linspace(0.0,5.0,100)
y = np.sin(2*np.pi*x)*np.exp(-x/3)

#
font = {'family': 'Calibri',
        'color':  'navy',
        'weight': 'normal',
        'size': 17,
         'alpha':0.2
        }
#
plt.plot(x,y,'--',color ='darkorange')
plt.title('Damped exponential decay', 
          fontdict=font,
          loc = 'left',
          y = 0.5)

```
