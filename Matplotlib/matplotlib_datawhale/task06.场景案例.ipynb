####  1.1  折线图 - Line chart

**折线图** 是一种基本图表，它具有以下特点：

* 折线图的数据通常是**有序排列** 
* 折线图表示的是数据的**趋势变化**
* 折线图由线段连接绘制

```python
# 折线图


# 创建数据
df = pd.DataFrame({'xdata':range(1,101),
                   'ydata':np.random.randn(100)})

# 设置基本样式

plt.figure(figsize=(10,8))
plt.style.use('seaborn-darkgrid')



#绘图

plt.plot('xdata','ydata',data=df,
         color = 'darkorange',
         alpha= 0.8,
        linestyle='-.',
        label = 'linestyle',)

plt.legend(loc='upper right',
           fontsize='x-large',
           frameon =True)

plt.title('Line chart')
```
