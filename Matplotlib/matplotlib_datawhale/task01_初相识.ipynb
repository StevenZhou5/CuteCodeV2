####  2、简单的绘图例子


* **plt.subplots(m,n,figsize=(a,b))** 

　　创建m x n的子图，图形size为 a x b 默认m=1，n=1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 生成2*2 子图

fig,ax = plt.subplots(2,2,figsize=(5,5)) 

ax[0,0].plot([1,2,3,4],[1,4,2,3])
ax[0,1].plot([1,2,3,4],[1,4,2,3])
ax[1,0].plot([1,2,3,4],[1,4,2,3])
ax[1,1].plot([1,2,3,4],[1,4,2,3])

plt.show()
```

```python
# 默认生成一个子图


fig,ax = plt.subplots(1,1)  # 创建一个包含一个axes的subplot
ax.plot([1,2,3,4],[1,4,2,3])

plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,2,3]) # 画图会自动创建figure
```

####  3、Figure的组成

figure理解为画板

axes/subplot 是画板上的画布

axis为画布上的坐标轴

tick坐标轴上的刻度

####  4、两种绘图接口

* 手动创建画布figure 子图subplot  轴域axes 坐标轴axis

* 直接绘图，绘图时会自动生成figure

```python
## 手动创建画布


x = np.linspace(0,2,100)

fig,ax = plt.subplots()
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quadra')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('simple plot')
ax.legend()  # 小图标
```

```python
## 直接画图


x = np.linspace(0,2,100)


plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadra')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('simple plot')
plt.legend()  # 小图标
```
