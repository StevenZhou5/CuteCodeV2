```python
import matplotlib.pyplot as plt
import numpy as np

# 这个设置才可以让Mac下可以显示中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

plt.rcParams['figure.figsize'] = (8.0,5.0) # 设置图像的显示区域
```

# 直方图

```python
# 随机生成(10000,)服从正态分布的数据
data = np.random.randn(1000)
plt.hist(data,bins=50,facecolor='b',edgecolor='r',alpha=0.8)
plt.xlabel("区间")
plt.ylabel("频数/频率")
plt.title("频数/频率的分布直方图")
plt.show()
```

```python
# 一个班级的成绩分布情况
data = np.random.randint(0,100,100)
bins = [0,60,80,100]
plt.hist(data,bins=bins,facecolor='blue',edgecolor='black',alpha=0.6)
plt.xlabel("成绩")
plt.ylabel("人数")
plt.title("成绩分布直方图")
plt.show()
```

```python
# 两个班级的成绩对比图
scoresC1 = np.random.randint(0,100,100)
scoresC2 = np.random.randint(0,100,100)
x = [scoresC1,scoresC2]
labels = ["班级A","班级B"]
bins = range(0,101,10)
plt.hist(x,bins=bins,color=['g','b'],
         histtype='bar',
         stacked=True, # 是否要堆叠起来
         rwidth=0.5, # 间隔：1百分百填充没有间隔
         label=labels)
plt.legend()
plt.show()
```

# 饼图

```python
labels = ["支付宝",'微信',"刷卡","现金"]
sizes = [20543,12042,4520,1200]
plt.pie(sizes,labels=labels,
        colors=['r','g','b','y'],
        explode=(0.1,0,0.2,0), # 让某一块突出去
        shadow=True,
        radius=1.5, # 缩放比例
        autopct='%1.1f%%', # 显示保留一位小数的百分比
       )
# plt.legend()
plt.title("支付方式")
plt.show()
```

# 3D 图

```python
import matplotlib.pyplot as plt
import numpy as np

# 这个设置才可以让Mac下可以显示中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

plt.rcParams['figure.figsize'] = (8.0,5.0) # 设置图像的显示区域
from mpl_toolkits.mplot3d import Axes3D # 需要导入3D图形包
fig = plt.figure() # 定义画布
ax = Axes3D(fig) # 将3D网格绘制在画布上

X = np.arange(-4,4,0.25) # 生成均匀分布的点
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y) # 用X和Y生成N*N的网格坐标
Z = X**2 + Y **2
print(X.shape,Y.shape,Z.shape)
print(X)
print(Y)
print(Z)
# ax.scatter(X,Y,Z) # 绘制点
ax.plot_surface(X,Y,Z,
                rstride=1, # 在行上面的跨度
                cstride =2, # 在列上面的跨度
                cmap=plt.get_cmap("rainbow"), # 用彩虹风格绘制
               ) # 绘制面
plt.show()
```
