```python
import numpy as np
import matplotlib.pyplot as plt
# 下面这些方法都没法解决mac下无法显示中文的问题
# plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置字体，解决中文无法显示问题
# plt.rcParams['font.sans-serif'] = ['STHeiti'] # 指定默认字体
# plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

# from pylab import mpl

# mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
# mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

# 这个设置才可以让Mac下可以显示中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

```python
# 查看matplot可以显示的字体有哪些
from matplotlib.font_manager import FontManager
fm = FontManager()
mat_fonts = set(f.name for f in fm.ttflist)
print(mat_fonts)
```

```python
# 绘制点
x1 = np.array([4])
y1 = np.array([6])
plt.scatter(x1,y1,alpha=0.5)
plt.title("中文标题")
plt.show()
```

```python
# 绘制线
x2 = np.array([3,8])
y2 = np.array([5,7])
plt.plot(x2,y2)
plt.show()
```

# 点和线绘制在一张图上，添加文本，设置样式

```python
plt.scatter(x1,y1,color='blue')
plt.plot(x2,y2,
         color='yellow',
         linewidth=3, # 设置线宽
         linestyle='--' # 设置显示样式
        )
plt.text(x1-0.5,y1+0.2,"文字说明",fontdict={"size":16,"color":'b'})
plt.show()
```

# 通过figure绘制多图

```python
plt.figure() # 代表下面的代码是一个画布
plt.scatter(x1,y1,color='blue')
plt.figure() # 再来一个画布
plt.plot(x2,y2,color='red',linestyle='--')
plt.show()
```

# 坐标刻度

```python
x_ticks = [1,2,3,5,7] # x轴上的坐标刻度
plt.xticks(x_ticks)
y_ticks = [1,2,3,5,7] # y轴坐标刻度
y_labels = ["差","普通","好","很好","非常好"]
plt.yticks(y_ticks,y_labels)
plt.plot([1,7],[1,7],color='g',linestyle='--')
plt.show()
```

# 坐标范围

```python
plt.xlim(0,8)
plt.ylim(0,7)
plt.scatter(x1,y1,color='b')
plt.plot(x2,y2,color='black')
plt.show()
```

# 坐标标签

```python
plt.xlabel("我是x轴",
           fontsize=12,#设置字体大小
          )
plt.ylabel("我是y轴",
           fontsize=15
          )
plt.show()
```

# legend显示图例

```python
plt.plot(x2,y2,color='b',label='蓝色')
plt.plot(x2+1,y2-2,color='r',linestyle="--",label='红色')
plt.legend(loc='lower right')
plt.show()
```

# Subplot 多图显示

```python
plt.subplot(2,2,1) # 两行两列的第一个位置
plt.scatter(x1,y1,color='r')
plt.subplot(2,2,2) # 第二个位置
plt.scatter(x1,y1,color='b')
plt.subplot(2,2,3)# 第三个位置
plt.plot(x2,y2,color='r')
plt.subplot(2,2,4) # 第四个位置
plt.plot(x2,y2,color='b')
plt.show()
```

```python
# 不规则的显示
plt.subplot(2,1,1) # 两行一列的第一个位置
plt.scatter(x1,y1,color='r')
plt.subplot(2,2,3) # 两行两列的第三个位置
plt.plot(x2,y2,color='r')
plt.subplot(2,2,4) # 两行两列的第四个位置
plt.plot(x2,y2,color='b')
plt.show()
```
