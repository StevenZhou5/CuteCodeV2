```python
# 深度优先遍历图
from collections import deque
graph = [(1,2),(2,4),(2,5),(4,8),(5,8),(1,3),(3,6),(3,7),(6,7)]
print("graph:",graph)

graph_dict = {}
for edge1,edge2 in graph:
    edge1_next_list = graph_dict.get(edge1,[])
    edge1_next_list.append(edge2)
    graph_dict[edge1] = edge1_next_list
    
    edge2_next_list = graph_dict.get(edge2,[])
    edge2_next_list.append(edge1)
    graph_dict[edge2] = edge2_next_list
print(graph_dict)

# 深度优先搜索，深(堆)栈
start = 1
stack = deque([graph_dict.get(start,[])])
print(stack)
visted = [1]
print(visted)
print()

while stack:
    print("*"*100)
    print(stack)
    cur_next_list = stack.pop()
    print("开始：",cur_next_list)
    if not cur_next_list: # 如果当前结点没有可访问的边了
        continue
    
    next_node = cur_next_list.pop(0)
    stack.append(cur_next_list) # 把剩余结点加进去
    if next_node in visted: # 如果已经访问过，那么直接访问此结点的另一条边
        continue
    
    visted.append(next_node)
    stack.append(graph_dict.get(next_node,[]))
    print("结束：",cur_next_list)

visted
```
