```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def reorderList(self, head: ListNode) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        if not head:
            return head
        # step1：找到中间结点
        step1_node = head
        step2_node = head.next
        while step2_node and step2_node.next:
            step1_node = step1_node.next
            step2_node = step2_node.next
            if not step2_node:
                break
            step2_node = step2_node.next

        # step2: 翻转后面一半的链表
        pre_node = None
        cur_node = step1_node.next
        step1_node.next = None
        while cur_node:
            tem = cur_node.next
            cur_node.next = pre_node
            pre_node = cur_node
            cur_node = tem
        # print(pre_node.next.next)
        
        # step3: 重组
        cur_pre_node = head
        cur_end_node = pre_node
        while cur_pre_node and cur_end_node:
            tem_pre = cur_pre_node.next
            tem_end = cur_end_node.next
            cur_pre_node.next = cur_end_node
            cur_end_node.next = tem_pre
            cur_pre_node = tem_pre
            cur_end_node = tem_end
            
        
```

```python
# 给定一个单链表，把所有的奇数节点和偶数节点分别排在一起。请注意，这里的奇数节点和偶数节点指的是节点编号的奇偶性，而不是节点的值的奇偶性。

# 请尝试使用原地算法完成。你的算法的空间复杂度应为 O(1)，时间复杂度应为 O(nodes)，nodes 为节点总数。

# 示例 1:

# 输入: 1->2->3->4->5->NULL
# 输出: 1->3->5->2->4->NULL
# 示例 2:

# 输入: 2->1->3->5->6->4->7->NULL 
# 输出: 2->3->6->7->1->5->4->NULL
# 说明:

# 应当保持奇数节点和偶数节点的相对顺序。
# 链表的第一个节点视为奇数节点，第二个节点视为偶数节点，以此类推。

# 来源：力扣（LeetCode）
# 链接：https://leetcode-cn.com/problems/odd-even-linked-list
# 著作权归领扣网络所有。商业转载请联系官方授权，非商业转载请注明出处。
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def oddEvenList(self, head: ListNode) -> ListNode:
        if not head:
            return head
        
        end_odd = head 
        start_even = head.next # 这个始终不能变，用于最后连接
        end_even = start_even
        while end_even:
            next_odd = end_even.next # 下一个奇数
            if not next_odd:
                break
            next_even = next_odd.next # 下一个偶数

            end_odd.next = next_odd # 上一个末尾奇数要连接到下一个奇数
            end_even.next = next_even # 上一个末尾偶数要连接到下一个偶数

            end_odd = next_odd # 更新奇数末尾
            end_even = next_even # 更新偶数末尾

        
        # 最后end_odd 要连接到start_even
        end_odd.next = start_even
        return head
```

```python
# 给你链表的头结点 head ，请将其按 升序 排列并返回 排序后的链表 。
# 进阶：
# 你可以在 O(n log n) 时间复杂度和常数级空间复杂度下，对链表进行排序吗？
# 示例 1：
# 输入：head = [4,2,1,3]
# 输出：[1,2,3,4]
# 示例 2：
# 输入：head = [-1,5,3,4,0]
# 输出：[-1,0,3,4,5]

# 来源：力扣（LeetCode）
# 链接：https://leetcode-cn.com/problems/sort-list
# 著作权归领扣网络所有。商业转载请联系官方授权，非商业转载请注明出处。
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def merge(self,node1,node2):
        sort_end = None
        next1 = node1 
        next2 = node2
        while next1 and next2:
            if next1.val <= next2.val: # 如果结点1的val小于等于
                if sort_end:
                    sort_end.next = next1
                sort_end = next1
                next1 = next1.next
            else:
                if sort_end:
                    sort_end.next = next2
                sort_end = next2
                next2 = next2.next
        sort_end.next = next1 if next1 else next2
        return node1 if node1.val <= node2.val else node2


    def sortList(self, head: ListNode) -> ListNode:
        if not head or not head.next:
            return head
        # 二分归并
        step1_node = head
        step2_node = head.next
        while step2_node and step2_node.next:
            step1_node = step1_node.next
            step2_node = step2_node.next
            if step2_node:
                step2_node = step2_node.next
        # print(step1_node.val)
        # print(step2_node.val)
        middle_node = step1_node.next
        step1_node.next = None
        sort1 = self.sortList(head)
        sort2 = self.sortList(middle_node)
        res = self.merge(sort1,sort2)
        return res
```
