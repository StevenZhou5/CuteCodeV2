# 基础位远算


## 注意异或运算^的一些特性
```python
x ^ 0 = x 
x ^ 1s = ~x # 1s = ~0 即所有位都为1的数,
x ^ (~x) = 1s
x ^ x = 0 
a ^ b = c => a ^ c = b, b ^ c = a #  可以用这种方式来交换（swap）a、b 的值
a ^ b ^ c = a ^ (b ^ c) = (a ^ b) ^ c # associative,满足结合率
a ^ b ^ c = a ^ c ^ b # 异或运算也满足交换率
```

````python
print("~0的值为：{0}".format(~0))
````

# 实战中常用位远算
```python
x & 1 == 1 OR == 0 # 判断x的奇偶，== 1 为奇数，== 0 为偶数；
x = x & (x - 1) # 清零最低位的1
x & -x # 得到最低位的 1； -x 的二进制 = x的二进制按位取反在+1
x = x & (~(1 << n)) # 仅把x的第n位置为0
x = x | (1 << n) # 仅把x的第n位置为1；
x = x ^ (1 << n) # 仅把x的第n位取反；常在状态压缩的回溯中使用，
x = x & (~0 << n) # 将x的最右边n位清零
(x >> n) & 1 # 获取x第n位的值（0 或者 1）
x = x & ((1 << n) - 1) # 将x的最高位至第n位(含)清零； 在N皇后和数独中都有用到
x = x & ((~0 >> n + 1) << n + 1) = x & (~((1 << n + 1) - 1)) # 将第n位置第0位(含)清零
```

````python
bin(31 & ((~0 >> 3 + 1) << 3 + 1))
````

````python
bin(31 & (~((1 << 3) - 1)))
````
