```python
import tensorflow as tf
import numpy as np
```

```python
f1_list = np.array([np.random.randint(1,100) for _ in range(100)])
print(f1_list)
print()
f2_list = np.array([np.random.randint(1,100) for _ in range(100)])
print(f2_list)
print()
label = f1_list + f2_list * 2
print(label)
```

```python
f1_input_layer = tf.keras.layers.Input(shape=(),name='f1_input')
flattened_f1_layer = tf.keras.layers.Flatten()(f1_input_layer)

f2_input_layer = tf.keras.layers.Input(shape=(),name='f2_input')
flattend_f2_layer = tf.keras.layers.Flatten()(f2_input_layer)

concatenate_layer = tf.keras.layers.concatenate([flattened_f1_layer,
                                                 flattend_f2_layer],
                                                axis=-1)
hidden_layer1 = tf.keras.layers.Dense(8,activation='relu')(concatenate_layer)
ouput_layer = tf.keras.layers.Dense(1)(hidden_layer1)

model = tf.keras.models.Model(inputs=[f1_input_layer,f2_input_layer],
                              outputs=ouput_layer)

model.compile(optimizer=tf.keras.optimizers.Adam(),loss='mse')
model.summary()
```

```python
history = model.fit({"f1_input":f1_list,
                     "f2_input":f2_list},
                    label,batch_size=32,epochs=1000)
```

```python
model.predict({"f2_input":np.array([5]),
               "f1_input":np.array([3])})
```
