```python
model = keras.models.Sequential([
    keras.layers.InputLayer(input_shape = [2]),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(64,activation="relu"),
    keras.layers.Dense(32,activation="relu"),
    keras.layers.Dense(1)
])

model.compile(loss='mse',
              optimizer = keras.optimizers.Adam(0.001),
#               optimizer =keras.optimizers.SGD(learning_rate=0.1),
              metrics = ["accuracy"])
model.summary()

history = model.fit(train_1,train_label,batch_size=128,epochs=100)
```

```python
train_x = [i for i in range(100) if i % 2 == 1]
train_y = [[2 * x + 5, 3 * x -6] for x in train_x]
train_y
```

```python
valid_x = [i for i in range(20) if i % 2 == 0]
valid_y = [[2 * x + 5, 3 * x -6] for x in valid_x]
valid_y
```

```python
model = keras.models.Sequential([
    keras.layers.InputLayer(input_shape = [1]),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(4,activation="relu"),
    keras.layers.Dense(2)
])

model.compile(loss='mse',
              optimizer = keras.optimizers.Adam(0.1),
              metrics = ["accuracy"])
model.summary()
```

```python
history = model.fit(train_x,
                    train_y,
                    epochs=1000,
                    validation_data=(valid_x,valid_y))
```

```python
model.evaluate([105,208],[[215,309],[421,618]])
```
