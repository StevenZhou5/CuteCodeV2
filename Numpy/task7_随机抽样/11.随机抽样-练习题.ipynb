## 2. 生成相应的数据

【知识点：随机抽样】

创建分别具有5000个数据的训练集（xi，y）和测试集（xi，y），其中xi在间隔（-10,10）上均匀随机分布。
为了使回归问题“真实”，大的均匀噪声分布在[-0.2,0.2]已添加到所有训练样本中，同时测试数据保持无噪声。

$$
f(x)=
\begin{cases}
sin(x)/x & \text{x not= 0 }\\
1& \text{x=0}
\end{cases}$$




```python
#生成训练集数据（xi,y）
import matplotlib.pyplot as plt


np.random.seed(2020)
train_x = np.random.uniform(low = -10,high=10,size=5000)
train_x = np.round(train_x,4)
train_y=[(np.sin(i)/i) if i !=0 else 1 for i in train_x]

# 
plt.plot(train_x,train_y)
plt.show()
```
